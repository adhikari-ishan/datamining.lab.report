```python
print("Ishan Adhikari")
# Drop StudentID column

# Encode categorical variables
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

# Convert ExtraClasses (Yes/No) into numbers
df["ExtraClasses"] = encoder.fit_transform(df["ExtraClasses"])

# Convert Result (Pass/Fail) into numbers
df["Result"] = encoder.fit_transform(df["Result"])

# Display updated dataset
df.head()
```

```python
print(df.columns)
```

```python
from sklearn.model_selection import train_test_split
print("----Ishan Adhikari----")
# Separate features (X) and target variable (y)
X = df.drop("Result", axis=1)
y = df["Result"]

# Split the dataset into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

# Display the shape of each dataset
print("Training Features:", X_train.shape)
print("Testing Features:", X_test.shape)
print("Training Labels:", y_train.shape)
print("Testing Labels:", y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
print("----Ishan Adhikari----")
# Decision Tree using Gini Index
dt_gini = DecisionTreeClassifier(
    criterion="gini",
    random_state=42
)

dt_gini.fit(X_train, y_train)

print("Decision Tree model trained using Gini Index.")
```

```python
# Decision Tree using Entropy
print("----Ishan Adhikari----")
dt_entropy = DecisionTreeClassifier(
    criterion="entropy",
    random_state=42
)

dt_entropy.fit(X_train, y_train)

print("Decision Tree model trained using Entropy.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
print("----Ishan Adhikari----")
# Predict using the Gini model
y_pred_gini = dt_gini.predict(X_test)

# Evaluation metrics
accuracy = accuracy_score(y_test, y_pred_gini)
precision = precision_score(y_test, y_pred_gini)
recall = recall_score(y_test, y_pred_gini)
f1 = f1_score(y_test, y_pred_gini)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-Score :", f1)
```

```python
# Predict using the Entropy model
y_pred_entropy = dt_entropy.predict(X_test)
print("----Ishan Adhikari----")
# Evaluation metrics
print("Accuracy :", accuracy_score(y_test, y_pred_entropy))
print("Precision:", precision_score(y_test, y_pred_entropy))
print("Recall   :", recall_score(y_test, y_pred_entropy))
print("F1-Score :", f1_score(y_test, y_pred_entropy))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
print("----Ishan Adhikari----")
# Confusion Matrix for Gini
cm = confusion_matrix(y_test, y_pred_gini)

ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Fail", "Pass"]
).plot(cmap="Blues")

plt.title("Confusion Matrix (Decision Tree - Gini)")
plt.show()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
print("----Ishan Adhikari----")
# Visualize the Decision Tree trained with Gini criterion
plt.figure(figsize=(15, 10))

plot_tree(
    dt_gini,
    feature_names=X.columns,
    class_names=["Fail", "Pass"],
    filled=True,
    rounded=True,
    fontsize=10
)

plt.title("Decision Tree Visualization (Gini)")
plt.show()
```

```python
plt.figure(figsize=(15, 10))

plot_tree(
    dt_entropy,
    feature_names=X.columns,
    class_names=["Fail", "Pass"],
    filled=True,
    rounded=True,
    fontsize=10
)

plt.title("Decision Tree Visualization (Entropy)")
plt.show()
```
